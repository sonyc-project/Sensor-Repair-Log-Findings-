### Rogers Facing Jay Street

Below is the data for the Rogers Facing Jay sensor. 

```python
rogers_facing_jay=pd.read_csv("./data/rogers_facing_jay_clean.csv")
```

I began my analysis by converting the csv data into a dataframe. This way, I am able analyse the data using pandas and matplotlib. A sample of the data that I converted is printed below. 

```python
print(rogers_facing_jay)
```

```python
rogers_facing_jay_cleaned = rogers_facing_jay.dropna(axis=0, how="any", thresh=None, subset=None, inplace=False)
```

I then cleaned the data using the build in dropna() function in pandas. This iterates through the data and removes any null or invalid datapoints. I then printed a sample of that analysis below. 

```python
print(rogers_facing_jay_cleaned)
```

```python
1464 - 592
```

After cleaning the data, pandas found 592 rows of usable data. This means that 872 rows were removed in the cleaning.

Below is the plot of the data before cleaning. The Y axis represents wifi signal quality out of 100% and the X representes time.

```python
ax1 = rogers_facing_jay.plot(figsize=(8,5), color='Red')
x_axis = ax1.axes.get_xaxis()
x_axis.set_visible(False)
ax1.set_ylabel("Wifi Signal Quality %")
ax1.set_title("Rogers Facing Jay")
```

This is the plot of the clean version.

```python
ax1_clean = rogers_facing_jay_cleaned.plot(figsize=(8,5), color='Red')
x_axis = ax1_clean.axes.get_xaxis()
x_axis.set_visible(False)
ax1_clean.set_ylabel("Wifi Signal Quality")
ax1_clean.set_title("Rogers Facing Jay Cleaned")
```

The cleaned data removes the gaps from the original data while still maintaining the same overall shape. This allows a clearer picture of the data's movement.

## Conclusion

Reliance on the NYU wifi network other potential fault point not directly related the the state of the sensor. A sensor could be in prefect condition, but if NYU's wifi fails, the sensor will ultimately fail. Adding cellular cards to these specific sensors and other sensors with a history of poor wifi signal decrease the chance outside wifi failures. Faults that occur while using cellular cards would be more diagnosable because then a componend of the sensor itself is failing. Now, if there is a wifi-related issue, it is harder to diagnose if it is one of the sensor's components failing or if is the overall network.
